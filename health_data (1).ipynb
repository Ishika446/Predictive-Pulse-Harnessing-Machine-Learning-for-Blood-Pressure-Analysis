```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
df=pd.read_csv("data.csv")
```

```python
df
```

```python
main=["Blood_Pressure_Abnormality","Level_of_Hemoglobin","Genetic_Pedigree_Coefficient","Age","BMI","Sex","Pregnancy","Smoking","Physical_activity","salt_content_in_the_diet","alcohol_consumption_per_day","Level_of_Stress","Chronic_kidney_disease","Adrenal_and_thyroid_disorders"]
```

```python
df_main=df[main]
```

```python
df_main
```

```python
df_main = df_main.fillna(0)
```

```python
df_main
```

```python
df_main['Pregnancy'] = df_main['Pregnancy'].astype(int) 
```

```python
df_main['alcohol_consumption_per_day'] = df_main['alcohol_consumption_per_day'].astype(int) 
```

```python
df_main
```

```python
gender_counts = df['Sex'].value_counts()
plt.figure(figsize=(6,6))
new_label = ['Men', 'Women'] 
plt.pie(gender_counts, labels=new_label, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#ff9999'])
plt.title('Gender Distribution')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
plt.show()
plt.savefig("pie_chart.png")
```

```python
sns.countplot(x='Smoking', hue='Level_of_Stress', data=df)

plt.title('Count plot of Smoking by Level of Stress')
plt.xlabel('Smoking')
plt.ylabel('Count')
plt.legend(title='Level of Stress')
plt.show()
plt.savefig("smoking vs level of stress.png")
```

```python
label_encoder = LabelEncoder()
for col in main:
    df_main[col] = label_encoder.fit_transform(df[col])
```

```python
df_main.head()
```

```python
select_col=['Age', 'Blood_Pressure_Abnormality', 'Level_of_Hemoglobin']
sns.pairplot(df_main[select_col])
plt.show()
plt.savefig("Multivariate analysis.png")
```

```python
plt.hist(df['Age'], bins=20, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
plt.savefig("Univariate analysis.png")
```
